The 2017 NBA playoffs were fun to watch for basketball fans, and Basketball Reference contains playoff data for each team that was in the playoffs. This is some exploratory data analysis on data combined for the top 8 teams from the NBA playoffs last year: Warriors, Cavaliers, Spurs, Celtics, Rockets, Wizards, Jazz, and Raptors. All the data is from https://www.basketball-reference.com/.

```python
#Import modules needed for this EDA
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import os
import re
from itertools import izip
import seaborn as sns
```

```python
#Get the current working directory
indir = os.getcwd()
csvs = []
#Iterate through each csv file that contains data in this directory 
for root, dirs, filenames in os.walk(indir):
    for f in filenames:
        if f.endswith(".csv"):
            csvs.append(f)
#print csvs

#Add a label column for the Team that a player plays for to each individual data frame
withTeams = []
for file in csvs:
    df = pd.read_csv(file)
    #keep the first 3 letters of the filename, which happen to be the team name
    df['Team'] = file[:3]
    withTeams.append(df)
    
#print withTeams
 
#Combine all of the NBA data together into one large data frame
allData = pd.concat(withTeams, ignore_index = True)
print allData.shape
allData.head(25)
```

The column name for the players is currently "Unnamed: 1" instead of Players, and the values in this column include the player names with the last parts of the URLs used to access player information on Basketball Reference's website, and these are both issues that need to be corrected. First, we can change the column name to Players.

```python
allData = allData.rename(columns = {"Unnamed: 1": "Players"})
allData.head(5)
```

Now that the column is properly named, we need to address the URL portion of the values inside the Players column. As it turns out, you need to escape the \ character in order to even identify its presence in the original data frame. Once that is done, we can remove the link information after the \ character and only keep the name of the player that we are interested in by using Python's split function and subsetting to keep the part of the result without the \ character.

```python
allData['Players'] = allData['Players'].str.split('\\').str[0]
allData.head(5)
```

In the data, the Games Played, or G category is the number of playoff games that a player played in, and they are different for players depending on how many games their team played in the playoffs. Thus, we should normalize the statistics, which are total amounts, to obtain per-game statistics to make the data more easy to compare amongst individual players who may not have played the same amount of playoff games.

```python
#let's see a list of all the variables to see which per game columns we want
df = allData

print df.columns  

'''Since the column names are not organized in a way such that percentages are separated from total values, 
    we need to manually specify the relevant columns we want
'''
relevantStats = ['MP','FG','FGA','3P','3PA','2P','2PA','FT','FTA','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PTS']

#Add the per-game data label

statsPG = []
for colName in relevantStats:
    statsPG.append(colName + 'perGame')
print statsPG
```

Now that the new column names are in place, let's actually calculate the per game statistics, and add those as columns to the existing data frame.

```python
#Using izip to iterate over two collections in a more memory efficient way than zip
for i, j in izip(relevantStats, statsPG):
    #perform the calculation and round to 2 decimal places
    df[j] = (df[i]/df.G).round(2)
df.head(10)
```

Which players averaged the most points per game?

```python
print df[['Players','PTSperGame']].sort_values('PTSperGame', ascending = False).head()
```

Which players averaged the most assists per game?

```python
print df[['Players','ASTperGame']].sort_values('ASTperGame', ascending = False).head()
```

```python
print df[['Team','PTSperGame']].groupby('Team').sum().sort_values('PTSperGame', ascending = False)
```

What is the correlation between different player statistics?

```python
#correlation for overall statistics 
df[['Age','MP','FG','FGA','3P','3PA','2P','2PA','FT','FTA','ORB','DRB','TRB','AST','STL','BLK','TOV','PF']].corr()
```

Heatmap for Per Game statistics

```python
#create a new data frame consisting of only columns with perGame stats
newdf = df.iloc[:,28:]
corr = newdf.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            cmap=sns.diverging_palette(255, 10, sep=10, as_cmap=True))
plt.show()
```

From the correlation matrix and the heatmap, there is a positive correlation between minutes and field goal attempts both on a total and per game basis.

The average age of players in the 2017 NBA playoffs was 27 years old. In the NBA, players who are over the age of 30 are sometimes considered old because at that point, they have usually been playing in the NBA for a long time and have a lot of wear and tear on their bodies as a result. Let's examine the count of minutes played for players above and below this age.

```python
plt.figure()
plt.style.use('fivethirtyeight')
plt.title("Minutes per game for players above and below age 30")
plt.xlabel("Average Minutes Per Game")
plt.ylabel("Count")
plt.ylim([0,20])
bins = np.linspace(0,48, 14)
plt.hist(df.MPperGame[df['Age'] > 30], bins = bins, alpha = 0.5, label = 'More than 30 years')
plt.hist(df.MPperGame[df['Age'] <= 30], bins = bins, alpha = 0.5, label= 'Less than 30 years')
plt.legend(loc='upper right')
plt.show()
```

From this histogram, we can see that very few players above the age of 30 play many minutes in the NBA playoffs at all. 
There appears to be only 1 player who averaged more than 40 minutes per playoff game who was above the age of 30. Let's find out who that was.

```python
print df[['Players','Age','MPperGame']][(df['MPperGame'] > 40)]
```

At age 32, LeBron James played the most playoff minutes per game in the NBA playoffs last year, and no other players played more than 40 minutes per playoff game last year.
